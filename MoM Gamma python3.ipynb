# Standard Gamma model: introduction

We consider the following Gamma model, with the two parameters $\alpha$ and $\beta$ unknown


$$
\bigg\{  Ga\big( \alpha, \beta   \big) ; \alpha > 0 , \ \beta > 0 \bigg\} 
$$

The Probability Density Function (PDF) of the Gamma distribution is given by

$$
f_{\alpha, \beta}(x) = \frac{  \beta^{\alpha}   }{ \Gamma(\alpha)  } x^{\alpha - 1}   e^{- \beta x} 
$$


The gamma function $\Gamma(\cdot)$, also called the Euler integral of the second kind is defined, for a nonnegative integer $\alpha$, as $\Gamma(\alpha) = \int\limits_{0}^{\infty} \ x^{\alpha -1}   e^{-x} dx.

# Method of Moments (MoM) estimation


\textbf{Method of Moments (MoM)}: If $ E[\varphi(X) ] = h(\theta)$, then we have that $\hat{\theta}_{MOM} = h^{-1}\bigg(  \frac{1}{n} \sum\limits_{i=1}^{n} \varphi(x_{i})  \bigg)$ is a method of moments estimator for $\theta$.

To obtain a \textbf{Method of Moments estimator} for the parameter vector $\boldsymbol{\theta} = ( \alpha \ , \  \beta)$ from a sample $x_{1},..., x_{n}$ of $n$ $i.i.d.$ realizations of a Gamma  distributed r.v $X$, we need the moments of $X$, given by 

$$ E[X] = \frac{\alpha}{\beta}  \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \  E[X^2] = \frac{\alpha(\alpha+1)}{\beta^{2}}$$


# Derivation of the first moment


$$
\begin{align*} 
& E[X] = \int\limits_{-\infty}^{+\infty} x \  f_{\alpha, \beta}(x)     \ dx =   \int\limits_{0}^{\infty} x \ \frac{  \beta^{\alpha}   }{ \Gamma(\alpha)  } x^{\alpha - 1}   e^{- \beta x}      \ dx  \\
& \ \ \ \ \ \ \  = \frac{  \beta^{\alpha}   }{ \Gamma(\alpha)  }           \int\limits_{0}^{\infty} x \  x^{\alpha - 1}   e^{- \beta x}      \ dx    \ \  \ (u= \beta x  	\Leftrightarrow  x= u/ \beta, \ \ \ du = \beta dx  \Leftrightarrow  dx= du / \beta ) \\
& \ \ \ \ \ \ \  = \frac{  \beta^{\alpha}   }{ \Gamma(\alpha)  }           \int\limits_{0}^{\infty}      \bigg(   \frac{u}{\beta}   \bigg)^{\alpha}   e^{- u}      \ \frac{du}{\beta}    \\          
& \ \ \ \ \ \ \  = \frac{  \beta^{\alpha}   }{ \Gamma(\alpha)  }         \frac{1}{\beta^{\alpha}}         \frac{1}{\beta}         \int\limits_{0}^{\infty}     u^{\alpha}   e^{- u}      \  du      \ \ \ \ \ \ \ \ \ \ \ \ \ \     \int\limits_{0}^{\infty}     x^{\alpha}   e^{- x}      \  dx  =  \Gamma(\alpha +1)  = \alpha  \Gamma(\alpha)    \\ 
& \ \ \ \  \ \ \  = \frac{  \beta^{\alpha}   }{ \Gamma(\alpha)  }         \frac{1}{\beta^{\alpha +1}}     \alpha  \Gamma(\alpha)   \\
& \ \ \ \  \ \ \  = \frac{\alpha}{\beta}  \ \ \ \ \ \ \ \  \  \\  
\end{align*}
$$

# Derivation of the second moment

$$
\begin{align*} 
& E[X^2] = \int\limits_{-\infty}^{+\infty} x^2 \  f_{\alpha, \beta}(x)     \ dx =   \int\limits_{0}^{\infty} x^2 \ \frac{  \beta^{\alpha}   }{ \Gamma(\alpha)  } x^{\alpha - 1}   e^{- \beta x}      \ dx  \\
& \ \ \ \ \ \ \  = \frac{  \beta^{\alpha}   }{ \Gamma(\alpha)  }           \int\limits_{0}^{\infty} x^2 \  x^{\alpha - 1}   e^{- \beta x}      \ dx    \ \ \ \ \ \ \ \ (u= \beta x  	\Leftrightarrow  x= u/ \beta, \ \ \ du = \beta dx  \Leftrightarrow  dx= du / \beta ) \\
& \ \ \ \ \ \ \  = \frac{  \beta^{\alpha}   }{ \Gamma(\alpha)  }           \int\limits_{0}^{\infty}      \bigg(   \frac{u}{\beta}   \bigg)^{\alpha + 1}   e^{- u}      \ \frac{du}{\beta}    \\       
& \ \ \ \ \ \ \  = \frac{  \beta^{\alpha}   }{ \Gamma(\alpha)  }         \frac{1}{\beta^{\alpha + 1}}         \frac{1}{\beta}         \int\limits_{0}^{\infty}     u^{\alpha}   e^{- u}      \  du      
\ \ \     \int\limits_{0}^{\infty}     x^{\alpha + 1}   e^{- x}      \  dx  =  \Gamma(\alpha + 2)   = (\alpha+1) \alpha \Gamma(\alpha)   \\ 
& \ \ \ \  \ \ \  = \frac{  \beta^{\alpha}   }{ \Gamma(\alpha)  }         \frac{1}{\beta^{\alpha +2}}     (\alpha+1) \alpha \Gamma(\alpha) \\
 & \ \ \ \  \ \ \ = \frac{(\alpha + 1) \alpha}{\beta^{2}}  \ \ \ \ \ \  \\  
\end{align*}
$$

# Gamma MoM estimators


So we set $\overline{x} = u$ and $\overline{x^{2}} = v$ and solve for $\alpha$ and $\beta$ in the following equation:

$$   h \bigg( \begin{matrix} \alpha\\ \beta \end{matrix}  \bigg) = \Bigg( \ \begin{matrix}   \alpha / \beta \\   \alpha(\alpha+1) / \beta^{2}  \end{matrix}  \Bigg)  = \Bigg( \ \begin{matrix}   u \\   v  \end{matrix}  \Bigg) 
$$

$$
 \begin{array}{lll}
  u = \alpha / \beta    \\
  v = \alpha(\alpha+1) / \beta^{2}  \ \ \  \ \ \Leftrightarrow \ \ \ \ \ \ \ \ v  =  u^{2}+ \frac{u}{\beta}   \ \ \  \ \ \Leftrightarrow \ \ \ \ \ \ \ \ v -  u^{2}= \frac{u}{\beta}   \ \ \  \ \ \Leftrightarrow \ \ \ \ \ \ \ \ \beta  = \frac{u}{v-u^{2}}\\
 \end{array}
$$


Then it is easy to show that  $\alpha = u \beta  \ \ \  \ \ \Leftrightarrow \ \ \ \ \ \ \ \ \alpha = \frac{u^{2}}{v-u^{2}}$

So we get as our Method of Moments estimators:

$$
\bigg( \frac{\overline{x}^{2} }{\overline{x^{2}} - \overline{x}^{2}},    \frac{\overline{x} }{\overline{x^{2}} - \overline{x}^{2}}     \bigg) =  \bigg(  \hat{\alpha}_{MOM}, \hat{\beta}_{MOM}   \bigg)
$$

# Example 1 - Lifetime of a product


The Gamma distribution is often used in reliability analysis to model time until failure kind of events, in particular Poisson distributed events. Suppose that we test the lifetime of a specific product on a sample of size $100$ and obtain the following data, in weeks (20 first entries of the dataset)

$$
\begin{align*}
x_{i} & = 17.49, 14.93, 16.01, 17.85,  7.37,  5.96, 12.27, 16.44, 28.34, 30.28, \\
& \ \ \ \ 10.66, 25.77, 17.21, 34.99, 28.53, 12.98,  9.32, 22.67, 12.55, 16.81
\end{align*}
$$


(i) \ \ From which Gamma distribution have the data been generated if we use the Method of Moment estimators for the parameters?


(ii) \ \ What are the cumulative hazard rates after 12, 24 and 36 weeks (or the total number of times we expect a failure for one product)?

# Example 1 - Estimation


The full dataset is available here:

The data have been generated from a $Ga\big( \alpha = 2, \beta = 0.08   \big)$ and computing the MoM estimators in R and Python, we can obtain a $Ga\big( \alpha = 1.900579, \beta = 0.07399452    \big)$ with R and a $Ga\big( \alpha = 1.900579, \beta = 0.073995   \big)$ with Python, so equivalent results up to rounding.




```python
import numpy as np
import pandas as pd

# function that implements the estimators above
def MoMgamma(x):
    n = len(x)
    sample_moment_1 = np.sum(x) / n
    sample_moment_2 = np.sum(x**2) / n

    alpha_mom = sample_moment_1**2 / (sample_moment_2 - sample_moment_1**2)
    beta_mom = sample_moment_1 / (sample_moment_2 - sample_moment_1**2)

    output = {
        'alpha_mom': alpha_mom,
        'beta_mom': beta_mom
    }

    return output

# read the dataset
x = pd.read_csv("C:/Users/julia/OneDrive/Desktop/github/37. momgamma/dataset.csv")

# perform MoM estimation
result = MoMgamma(x)
print(result)
```
